```python
def list_patient_names(root):

    patient_list = sorted([item for item in os.listdir(root) 
                           if os.path.isdir(os.path.join(root, item))],
                           key = lambda x: x[4:])
    
    return patient_list
```

```python
def list_trial_files(root, patient_name):
    
    patient_folder = root + patient_name + "/"
    
    trial_names = sorted([f for f in os.listdir(patient_folder) 
                          if os.path.isfile(os.path.join(patient_folder, f))])
    
    return trial_names
```

```python
root = os.getcwd() + '/'
```

```python
data_path = root + 'eeg_full/'
```

```python
patient_list = list_patient_names(data_path)
```

```python
headers = {}
```

```python
for p in patient_list:

    trial_names = list_trial_files(data_path, p)

    for t in trial_names:

        print(t)

        txtfile = data_path + p + "/" + t

        lines = []

        with gzip.open(txtfile,'rt') as f:
          for line in f:
            lines.append(line)

        try:
          if('err' in lines[3].split(' ')):
            headers[t] = lines[3]

        except:
          headers[t] = 'empty'
```

```python
len(headers)
```

```python
err_files = list(headers.keys())
```

```python
textfile = open(data_path+"err_files.txt", "w")
for element in err_files:
    textfile.write(element + "\n")
textfile.close()
```
